```python
import requests
from lxml import etree
import pandas as pd

pd.set_option('expand_frame_repr', False)
pd.set_option('display.max_rows', 1000)

import warnings

warnings.filterwarnings("ignore")

# url = 'https://www.baidu.com/'

# 抓取网址
url = 'https://tiyu.baidu.com/tokyoly/delegation/8705/tab/%E5%A5%96%E7%89%8C%E6%98%8E%E7%BB%86/current/0'
# https://tiyu.baidu.com/tokyoly/delegation/8705/tab/%E5%A5%96%E7%89%8C%E6%98%8E%E7%BB%86/current/0
```

```python
# 提取网页
txt = requests.get(url).text
txt
```

```python
html = etree.HTML(txt)
```

```python
# 名字 //*[@id="medal0"]/div[2]/b-lazyload/a[1]/div[2]/div[1]
#     //*[@id="medal0"]/div[2]/b-lazyload/a[1]/div[2]/div[1]
#     //*[@id="medal0"]/div[2]/b-lazyload/div[1]/div[2]/div[1] 
#    //*[@id="medal0"]/div[2]/b-lazyload/div[1]/div[2]/div[1]
    

# 项目 //*[@id="medal0"]/div[2]/b-lazyload/a[1]/div[2]/div[2]/span[1]
#     //*[@id="medal0"]/div[2]/b-lazyload/div[1]/div[2]/div[2]/span[1]

# //*[@id="medal0"]/div[2]/b-lazyload/div[1]/div[3]/div[1]/span 奖牌名次
# //*[@id="medal0"]/div[2]/b-lazyload/a[1]/div[3]/div[1]/span

# //*[@id="medal0"]/div[2]/b-lazyload/div[1]/div[3]/div[2] 获得日期
```

```python
name_test = html.xpath('''//*[@id="medal0"]/div[2]/b-lazyload/div[3]/div[2]/div[1]/text()''')[0]
name_test
```

```python
names=html.xpath('''//*[@id="medal0"]/div[2]/b-lazyload/.//div[2]/div[1]/text()''')
names[:3]
```

```python
projects = html.xpath('''//*[@id="medal0"]/div[2]/b-lazyload/.//div[2]/div[2]/span/text()''')
rankings = html.xpath('''//*[@id="medal0"]/div[2]/b-lazyload/.//div[3]/div[1]/span/text()''')
dates = html.xpath('''//*[@id="medal0"]/div[2]/b-lazyload/.//div[3]/div[2]/text()''')
```

```python
dates[:4]
```

```python
df=pd.DataFrame([names,projects,rankings,dates],index=["姓名","项目","顺序","日期"]).T
df
```

```python
df=df[df["顺序"].notnull()]
df
```

```python
df.loc[4,"日期"]
```

```python
df["日期"]=df["日期"].apply(lambda x:None if x ==' ' else x).fillna(method="bfill")
df.head(7)
```

```python
df.reset_index(inplace=True)
df.sort_values("index",ascending=0,inplace=True)
del df["index"]
df.head(8)
```
